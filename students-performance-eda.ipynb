```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
df=pd.read_csv('../input/student-performance/StudentsPerformance.csv')
```

```python
#reading top 5 lines
df.head()
```

```python
#looking for null values
df.isna().sum()
```

```python
#no.of female and male students
students=df['gender'].value_counts()
print(students)
print('\n')
#no.of students each race/ethnicity has
race=df['race/ethnicity'].value_counts()
print(race)
print('\n')
#looking at parents education
parented=df['parental level of education'].value_counts()
print(parented)
print('\n')
#no.of students in different segments of lunch types
lunch=df['lunch'].value_counts()
print(lunch)
print('\n')
#type of test preparaion courses
test=df['test preparation course'].value_counts()
print(test)
```

```python
race.to_dict()
parented.to_dict()
#pir charts of race and parents education
plt.figure(figsize=(12,12))
plt.subplot(1,2,1)
plt.pie(race,labels=race.keys(),autopct='%0.1f%%')
plt.subplot(1,2,2)
plt.pie(parented,labels=parented.keys(),autopct='%0.1f%%')
plt.show()
```

### As we have math,reading and writing scores, combining them will give overall performance of a student in assesment

```python
df['Total_score']=(df['math score']+df['reading score']+df['writing score'])/3
df
```

#### with total score we can compare how the performance of a students varies with the type of preparation course and other parameters

```python
plt.bar(df['race/ethnicity'], df['Total_score'])
```

```python
plt.figure(figsize=(8,8))
plt.bar(df['parental level of education'],df['Total_score'])
```

#### from the above graphs we can see that race and parental education doesn't have much affect on student's overall performance. Now let's see how these scores are distributed between male and female students

```python
y=df.loc[(df['gender']=='female')&df['Total_score']]
x=df.loc[(df['gender']=='male')&df['Total_score']]
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.distplot(y['Total_score'],label='Total_marks distribution for female students')
plt.legend()
plt.subplot(1,2,2)
sns.distplot(x['Total_score'],color='g',label='Total_score distribution for male students')
plt.legend()
```

```python
sns.relplot(x=df['math score'],y=df['reading score'],hue=df['test preparation course'])
sns.relplot(x=df['writing score'],y=df['reading score'],hue=df['test preparation course'])
sns.relplot(x=df['math score'],y=df['writing score'],hue=df['test preparation course'])
```
